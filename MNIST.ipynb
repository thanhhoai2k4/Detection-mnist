```python
# 1. import các thư viện
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten
from keras.layers import Conv2D, MaxPooling2D
from keras.utils import np_utils # chuyển y sang dạng one hot encoding
from keras.datasets import mnist
from keras.models import load_model
import cv2
```

```python
# 2. Load dữ liệu Mnist
(X_train, y_train), (X_test, y_test) = mnist.load_data()
X_val, y_val = X_train[50000:60000,:] , y_train[50000:60000]
X_train, y_train = X_train[:50000,:], y_train[:50000]
```

```python
# 3. reshape Dử liệu về đúng kích thước yêu cầu
X_train = X_train.reshape(X_train.shape[0],28,28,1)
X_val = X_val.reshape(X_val.shape[0],28,28,1)
X_test = X_test.reshape(X_test.shape[0],28,28,1)
```

```python
# 4. one hot encoding label y
Y_train = np_utils.to_categorical(y_train,10)
Y_val = np_utils.to_categorical(y_val,10)
Y_test = np_utils.to_categorical(y_test,10)
```

```python
# 5. Định nghĩa model
model = Sequential()
```

```python
model.add(Conv2D(32,(3,3),activation ='relu', input_shape=(28,28,1)))
```

```python
model.add(Conv2D(32,(3,3),activation ='relu'))
```

```python
model.add(MaxPooling2D(pool_size = (2,2)))
```

```python
model.add(Flatten())
```

```python
model.add(Dense(10,activation="softmax"))
```

```python
# Display the model's architecture
model.summary()
```

```python
# 6. compile model
model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics='accuracy')
```

```python
# 7. Thực hiện train model
H = model.fit(X_train,Y_train,validation_data=(X_val,Y_val),
              batch_size=32, epochs = 10, verbose = 1)
```

```python
# 8. Đánh giá model với dử liệu test
scores = model.evaluate(X_test,Y_test)
print("Loss: " , scores[0])
print("Accuracy: ", scores[1])
```

```python
# 9. vẽ đồ thị Loss va acc trong traning set va validation set
fig = plt.figure()
numofEpoch = 10

plt.plot(np.arange(0,numofEpoch),H.history["loss"], label = "traning loss")
plt.plot(np.arange(0,numofEpoch),H.history["val_loss"], label = "validation loss")
plt.plot(np.arange(0,numofEpoch),H.history["accuracy"], label = "accuracy")
plt.plot(np.arange(0,numofEpoch),H.history["val_accuracy"], label = "validation accuracy")
plt.xlabel("Epoch")
plt.title("Loss || Accurqacy")
plt.legend()

```

```python
# 10. Dự đoán ảnh 
plt.imshow(X_test[0].reshape(28,28),cmap="gray")


y_predict = model.predict(X_test[0].reshape(1,28,28,1))
print("DU DOAN: ", np.argmax(y_predict))
```
